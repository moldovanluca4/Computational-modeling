## 2A. Random walk with Absorbing Boundaries

#### Introduction/Theory

This variant models a particle in a finite domain where reaching the edge results in the particle being "removed" or "absorbed." It is used to study first-passage times and survival probabilities.

##### Theory and Formulas

1D Random Walk between x=0 and x=L:

- $\textbf{Absorbing Condition}$ - if particle position $x_i \leq 0$ or $x_i \geq L$ teh walk terminates here.

- $\textbf{Survival Probability}$ - probability that a walker has not been absorbed by time t, for large values of t the probability drops exponentially: 

$S(t) \propto e^{-t/\tau}$           where $\tau$ is the decay in time related to the diffusion constant D

- $\textbf{Mean First-Passage Time}$ - The avg time T it takes for a particle starting at a point $x_0$ to reach either boundary(0 or L):

$<T> = \frac{x_0 * (L - x_0)}{2*D}$

#### Numerical Methods

1) Simulation loop - we have W different walkers, and run them independent and track theire position until tha condition $x \in (0,L)$ becomes false


2) Collect Data - record the time step $N_{exit}$ at which the walkers get absorbed and create a histogram based on this to visualize the distribution of exit times.


3) Average - arithmetic mean of all $N_{exit}$






```python
#code for 2A 

import numpy

left_boundary = [4, 8, 16, 32, 64]
right_boundary = 50
x0 = 0
a = 1

exit_time = numpy.array([])
results = {}

number_of_walkers = 1000

for j in left_boundary:
    for i in range(number_of_walkers):
        current_x = x0
        steps = 0
    
        while -j < current_x < +j:
            r = numpy.random.rand()

            if r < 0.5:
                current_x = current_x + a
            else:
                current_x = current_x - a

            steps += 1
        
        exit_time = numpy.append(exit_time, steps)

        
    T_mean = numpy.mean(exit_time)
    results[j] = T_mean


print(results)




    

```

## 2B. Self-Avoiding Walk in 2D

The SAW is a path on a lattice that does not visit the same point more than once.

##### Theory and Formulas

What differs from the standard random walk is that the SAW has to memorise the previous path in order to avoid it => faster expansion in space

- $\textbf{SAW Constraint}$ - a step to a new position (x,y) is valid only if this new position is different from the previous positions

- $\textbf{Scaling Law}$ - the root mean square displacement: $\sqrt{<R^2>} \approx N^v$  - for a standard random walk in 2D: $v=0.5$, for a SAW in 2D: $v=0.75$

#### Numerical Methods

1) Lattice representation - 2D grid to track the path

2) Step execution - at each step theres 4 possibilities up, down, left, right choose randomly

3) Sampling strategies - if the walker hits itself discard the entire walk and start again (simple sampling method - could inefficient for large N)

- instead of equal probabilities, weight the stpes by the number of available free neighbours to prevent early      termination (Rosenbluth Method)

4) Analysis - calculate $R^2$ for many completed walk of length N and use a log-log plot to det the exp $v$ from the slope
  




```python
#code for 2B

import numpy 
import random

walk_length = [5, 10, 15, 20, 25, 30]
results = {}

for j in walk_length:
    succesfull_walk = []
    while len(succesfull_walk) < 100:
        path_is_visited = True
        visited = {(0,0)}
        x = 0
        y = 0 
        for i in range(j):
            neighbors = [(x+1, y), (x-1,y), (x, y+1), (x, y-1)]
            valid_neighbors = [n for n in neighbors if n not in visited]

            if not valid_neighbors:
                path_is_visited = False
                break
            x, y = random.choice(valid_neighbors)
            visited.add((x,y))
    
        if path_is_visited:
            r_square = x**2 + y**2
            succesfull_walk.append(r_square)


    results[tuple(walk_length)] = numpy.mean(succesfull_walk)

print(results)


```
